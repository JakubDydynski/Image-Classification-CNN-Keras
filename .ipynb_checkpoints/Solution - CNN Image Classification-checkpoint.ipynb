# Import Libraries

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

# Load Dataset

```python
X_train = np.loadtxt('input.csv', delimiter = ',')
Y_train = np.loadtxt('labels.csv', delimiter = ',')

X_test = np.loadtxt('input_test.csv', delimiter = ',')
Y_test = np.loadtxt('labels_test.csv', delimiter = ',')
```

```python
X_train = X_train.reshape(len(X_train), 100, 100, 3)
Y_train = Y_train.reshape(len(Y_train), 1)

X_test = X_test.reshape(len(X_test), 100, 100, 3)
Y_test = Y_test.reshape(len(Y_test), 1)

X_train = X_train/255.0
X_test = X_test/255.0
```

```python
print("Shape of X_train: ", X_train.shape)
print("Shape of Y_train: ", Y_train.shape)
print("Shape of X_test: ", X_test.shape)
print("Shape of Y_test: ", Y_test.shape)
```

```python
idx = random.randint(0, len(X_train))
plt.imshow(X_train[idx, :])
plt.show()
```

# Define a Network

```python
# Sequential - linear stack of layers. Output of layer n-1 is input of layer n.
model = Sequential([
    #layer 1 - convolutional, 32 filters of size 3x3, relu activation function, input shape of data  
    Conv2D(32, (3,3), activation = 'relu', input_shape = (100, 100, 3)),
    #layer 2 - MaxPooling with filter size 2x2
    MaxPooling2D((2,2)),
    
    #layer 3
    Conv2D(32, (3,3), activation = 'relu'),
    #layer 4
    MaxPooling2D((2,2)),
    
    #layer 5 - flattening layer
    Flatten(),
    
    #layer 6 - dense - fully connected layers, with  64 neurons and relu activation function
    Dense(64, activation = 'relu'),
    #layer 7 - dense - final layer with only 1 output neuron
    Dense(1, activation = 'sigmoid')
])
```

```python
model = Sequential()

model.add(Conv2D(32, (3,3), activation = 'relu', input_shape = (100, 100, 3)))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(32, (3,3), activation = 'relu'))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())
model.add(Dense(64, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

# Compile Network

```python
# loss - function used to measure performance during training
# optimizer - optimization algorithm used to update model's weights
# metrics - evaluation metrics to be computed
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

# Fit Network

```python
# x_train
# y_train
# epochs - number of iteration
# batch_size - number of samples in "processing unit" - every batch weights are updated
model.fit(X_train, Y_train, epochs = 6, batch_size = 64)
```

# Evaluate Network

```python
model.evaluate(X_test, Y_test)
```

# Make Predictions

```python
idx2 = random.randint(0, len(Y_test))
plt.imshow(X_test[idx2, :])
plt.show()

y_pred = model.predict(X_test[idx2, :].reshape(1, 100, 100, 3))
y_pred = float(y_pred[0])
print(y_pred)
print(f"This image is {100 * (1 - y_pred):.2f}% dog and {100 * y_pred:.2f}% cat.")


# binary classification
y_pred = y_pred > 0.5




if(y_pred == 0):
    pred = 'dog'
else:
    pred = 'cat'
    
print("Our model says it is a :", pred)
```
